```python
import matplotlib.pyplot as plt
from matplotlib import cycler
plt.style.use('classic')
import numpy as np

%matplotlib inline
```

```python
# there are ways to change the overall style of the plot all at onece
# we will sart with the classic style
x = np.random.randn(1000)
plt.hist(x)
```

```python
# adjusting this by hand can be tedious
# use a gray background
fig = plt.figure(facecolor="white")
ax = plt.axes(facecolor="#E6E6E6")
ax.set_axisbelow(True)

# draw solid white gridlines
plt.grid(color="w", linestyle="solid")

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors="gray", direction="out")
for tick in ax.get_xticklabels():
    tick.set_color("gray")
for tick in ax.get_yticklabels():
    tick.set_color("gray")

# control face and edge color of histogram
ax.hist(x, edgecolor="#E6E6E6", color="#EE6666")
```

```python
# each time matplotlib loads, it defines a runtime configuration containing the default styles for every plot element you create it can be adjusted using plt.rc
colors = cycler(
    "color", ["#EE6666", "#3388BB", "#9988DD", "#EECC55", "#88BB44", "#FFBBBB"]
)
plt.rc("figure", facecolor="white")
plt.rc(
    "axes",
    facecolor="#E6E6E6",
    edgecolor="none",
    axisbelow=True,
    grid=True,
    prop_cycle=colors,
)
plt.rc("grid", color="w", linestyle="solid")
plt.rc("xtick", direction="out", color="gray")
plt.rc("ytick", direction="out", color="gray")
plt.rc("patch", edgecolor="#E6E6E6")
plt.rc("lines", linewidth=2)
```

```python
# now that these are change we can plot the same plot as above
plt.hist(x)
```

```python
# the standard way to switch to a style is to call plt.style.use()
# this will change the style for the rest of the session
# alternatively you can use a context manager to revert back to the default style
# to demonstrate this we will create a quick function to create a plot
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(["a", "b", "c"], loc="lower left")
```

```python
with plt.style.context("default"):
    hist_and_lines()
```

```python
# the fivethirtyeight style mimics the graphics found on the popular website
with plt.style.context("fivethirtyeight"):
    hist_and_lines()
```

```python
# ggplot is based on the popular plotting package of the same name for the R programming language
with plt.style.context("ggplot"):
    hist_and_lines()
```

```python
# there uis also a style based on bayesian methods for hackers style
with plt.style.context("bmh"):
    hist_and_lines()
```

```python
# dark_background gives us a dark background with contrasting colors
with plt.style.context("dark_background"):
    hist_and_lines()
```

```python
# there is also a style based on grayscale
with plt.style.context("grayscale"):
    hist_and_lines()
```

```python
# there is also a style based on seaborn
with plt.style.context("seaborn-whitegrid"):
    hist_and_lines()
```
